```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
sales_csv = "Resources/vgsales.csv"
sales_df = pd.read_csv(sales_csv)
sales_df.head()
```

```python
rating_csv = "Resources/Video_games_esrb_rating.csv"
rating_df = pd.read_csv(rating_csv)
rating_df.head()
```

```python
clean_rating_df = rating_df[['title', 'esrb_rating']]
clean_rating_df
```

```python
clean_rating_df = clean_rating_df.rename(columns={'title': 'Name'})
clean_rating_df
```

```python
merge_df = pd.merge(sales_df, clean_rating_df, how="inner", on=['Name', 'Name'])
merge_df.head(20)
```

```python
merge_df = merge_df.drop_duplicates(subset=['Name'], keep='first')
merge_df
```

```python
merge_df = merge_df[['Rank', 'Name', 'Platform', 'Genre', 'Global_Sales', 'esrb_rating']]
merge_df
```

```python
merge_df = merge_df.rename(columns={'esrb_rating': 'ESRB_Rating'})
merge_df
```

```python
merge_df = merge_df.reset_index(drop=True)
merge_df
```

```python
merge_df = merge_df.rename(columns={'Rank': 'rank', 'Name': 'name', 'Platform': 'platform', 'Genre': 'genre', 'Global_Sales': 'global_sales','ESRB_Rating': 'esrb_rating'})
merge_df
```

```python
rds_connection_string = "postgres:postgres@localhost:5432/videogame_db"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
merge_df.to_sql(name='sales', con=engine, if_exists='append', index=True)
```

```python
pd.read_sql_query('select * from sales', con=engine).head()
```
